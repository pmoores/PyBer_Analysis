```python
# Import matplotlib dependency as "plt"
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using the MATLAB method
# plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() method - object-oriented programming
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend using MATLAB method
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend.
plt.legend()
```

```python
# SKILL DRILL - 5.1.4 - Using the ax_plot method (OOP) create a plot and add formatting as per instructions
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')

ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

ax.set_title("PyBer Fare by Month")

ax.legend("Boston")

ax.grid()
```

```python
# 5.1.6 - Add data to use with bar chart
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Add data to use with scatter plots

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatter plot using plt.plot()
plt.plot(x_axis, y_axis, 'o')
```

```python
# SKILL DRILL 5.1.7 - Create a scatter plot using the plt-scatter() method and switch axes so that Fare($) is on x-axis
plt.scatter(y_axis, x_axis, color='red', label='Chicago')

# Invert y axis so that Jan is on top
plt.gca().invert_yaxis()

# Add a title
plt.title("PyBer Fare by Month")

# Add axes labels
plt.ylabel("Date")
plt.xlabel("Fare($)")
              
# Add a limit for x axis data
plt.xlim([0, 50])

# Add a legend
plt.legend()
```

```python
# Create a bubble chart using the MATLAB method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Adjust the size of the markers by multiplying the data by any value (in this case 3)
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
# Use the new y axis for the s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Adjust the size of the markers using Python's list comprehension technique
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a scatter plot using the object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# SKILL DRILL 5.1.7 Create a bubble chart using object-oriented programming
fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c="Skyblue", alpha=0.2, label="Boston", linewidth=2, edgecolors='black')

# Add a title
ax.set_title("PyBer Fare by Month")

# Add axis labels
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')

ax.legend("Boston")

# Invert y axis so that Jan is on top
plt.gca().invert_yaxis()

# Add a limit of 0 to 50 for x-axis data
ax.set_xlim([0, 50])
```
